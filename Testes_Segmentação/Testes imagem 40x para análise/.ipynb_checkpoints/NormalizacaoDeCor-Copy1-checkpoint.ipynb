```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# carregar a imagem
img = cv2.imread('Carcinoma_40x.jpg')
img2 = cv2.imread('Carcinoma_40x.jpg')
img3 = cv2.imread('Carcinoma_40x.jpg')
img4 = cv2.imread('Carcinoma_40x.jpg')
img_preprocessada = cv2.imread('Carcinoma_40x.jpg')
```

```python
#Normalização de Cor
# converter para o espaço de cor RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# calcular a média e desvio padrão dos canais de cores
mean, std = cv2.meanStdDev(img)
mean = np.reshape(mean, (3,))
std = np.reshape(std, (3,))

# aplicar a normalização de cores
img = (img - mean) / std

plt.imshow(img)
```

```python
# Normaliza as cores por histograma
img_preprocessada = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_preprocessada[:,:,0] = clahe.apply(img_preprocessada[:,:,0])
img_preprocessada[:,:,1] = clahe.apply(img_preprocessada[:,:,1])
img_preprocessada[:,:,2] = clahe.apply(img_preprocessada[:,:,2])
```

```python
#Redução de Ruído com filtro gaussiano

# converter para escala de cinza
gray_img_preprocessada = cv2.cvtColor(img_preprocessada, cv2.COLOR_BGR2GRAY)

# aplicar filtro gaussiano para redução de ruído
gaussian = cv2.GaussianBlur(gray_img_preprocessada, (5, 5), 0)

#plt.imshow(gaussian)
```

```python
# Remoção do efeito da variação com técnica de correção de brilho e contraste 

# aplicar correção de brilho e contraste
alpha = 1.5  # ajuste de contraste (1.0 = nenhum ajuste)
beta = 50   # ajuste de brilho (0 = nenhum ajuste)
adjusted = cv2.convertScaleAbs(img3, alpha=alpha, beta=beta)

plt.imshow(adjusted)
```

```python
from skimage import exposure
from skimage.filters import unsharp_mask

# aplicar equalização de histograma
img_eq = exposure.equalize_hist(img4)

# aplicar normalização de textura
img_norm = unsharp_mask(img_eq, radius=5, amount=2)

plt.imshow(img_norm)
```
