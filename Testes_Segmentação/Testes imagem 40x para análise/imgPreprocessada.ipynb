```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# carregar a imagem
img = cv2.imread('Carcinoma_40x.jpg')
```

```python
# Normaliza as cores por histograma
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img[:,:,0] = clahe.apply(img[:,:,0])
img[:,:,1] = clahe.apply(img[:,:,1])
img[:,:,2] = clahe.apply(img[:,:,2])
```

```python
#Redução de Ruído com filtro gaussiano

# converter para escala de cinza
gray_img_preprocessada = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# aplicar filtro gaussiano para redução de ruído
gaussian = cv2.GaussianBlur(gray_img_preprocessada, (5, 5), 0)

#plt.imshow(gaussian)
```

```python
# Remoção do efeito da variação com técnica de correção de brilho e contraste 

# aplicar correção de brilho e contraste
alpha = 1.5  # ajuste de contraste (1.0 = nenhum ajuste)
beta = 50   # ajuste de brilho (0 = nenhum ajuste)
adjusted = cv2.convertScaleAbs(gaussian, alpha=alpha, beta=beta)

plt.imshow(adjusted)
```
